Given demand and supply modeled using Bipartite Graph with vertices marking objects and properties of individuals such as demand by person (1) for object (A) at time, location and cost [ object variables are weightings for the edges], and supplier with by person (2) for object (A) at time, location and cost. 

in this setting we can have instances where the market can not be served, due to unmatched demand and supply, which are instances in which the matching criteria hasn't been matched, thus leaving the participants to be active demand and supply participants, when a match exists between customers object (A:1) and suppliers object (A:1) - suppliers and customers objects are indexed based on the quantity - if participants have number of items left they can still participate in the market, but else they wont participate in the market until atleast 1 index of items to be borrowed or leased out is available. 

The primary interest is the demand and supply that can actively participate on the network.
